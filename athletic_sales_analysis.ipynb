```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis=0, join='inner').reset_index(drop=True)
combined_df.head()
```

```python
# Check if any values are null.
combined_df.isnull().sum()
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df.reset_index(inplace=True)
combined_df['invoice_date']= pd.to_datetime(combined_df['invoice_date'], errors='coerce')
combined_df.set_index('invoice_date', inplace=True)
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
products_df = combined_df.groupby(['region', 'state', 'city'])['units_sold'].agg(Total_Products_Sold=("sum")).reset_index()

# Rename the sum to "Total_Products_Sold".

# Show the top 5 results.
top_5 = products_df.sort_values(by=(["Total_Products_Sold"]),ascending=False).head()
top_5
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_df = combined_df.pivot_table(
    values='units_sold',
    index=['region', 'state', 'city'],
    aggfunc='sum'
).reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
top_5 = pivot_df.sort_values(by=(["Total_Products_Sold"]),ascending=False).head()
top_5
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_df = combined_df.groupby(['region', 'state', 'city'])['total_sales'].agg(Total_Sales=("sum")).reset_index()
# Rename the "total_sales" column to "Total Sales"
total_sales_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
top_5 = total_sales_df.sort_values(by=(["Total_Sales"]),ascending=False).head()
top_5
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_df = combined_df.pivot_table(
    values='total_sales',
    index=['region', 'state', 'city'],
    aggfunc='sum'
).reset_index()


# Optional: Rename the "total_sales" column to "Total Sales"
pivot_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
top_5 = total_sales_df.sort_values(by=(["Total_Sales"]),ascending=False).head()
top_5
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
sales_df = combined_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].agg(Total_Sales=("sum")).reset_index()

# Rename the "total_sales" column to "Total Sales"
sales_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
top_5 = sales_df.sort_values(by=(["Total_Sales"]),ascending=False).head()
top_5
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_df = combined_df.pivot_table(
    values='total_sales',
    index=['retailer', 'region', 'state', 'city'],
    aggfunc='sum'
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
top_5 = total_sales_df.sort_values(by=(["Total_Sales"]),ascending=False).head()
top_5
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales_df = combined_df[combined_df["product"].str.contains("Women's Athletic Footwear")]
womens_footwear_sales_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_sold_df = womens_footwear_sales_df.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].agg(units_sold=("sum")).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sold_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
top_5 = womens_footwear_sold_df.sort_values(by=(["Womens_Footwear_Units_Sold"]),ascending=False).head()
top_5
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_df = combined_df.pivot_table(
    values='units_sold',
    index=['retailer', 'region', 'state', 'city'],
    aggfunc='sum'
).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_df.rename(columns={'units_sold': 'Womens Footwear Units Sold'}, inplace=True)

# Show the top 5 results.
top_5 = womens_footwear_sold_df.sort_values(by=(["Womens_Footwear_Units_Sold"]),ascending=False).head()
top_5

```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
sales_pivot_sum = pd.pivot_table(combined_df,
    values='total_sales',
    index=['invoice_date'],
    aggfunc='sum'
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
sales_pivot_sum.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the table.
sales_pivot_sum

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = sales_pivot_sum.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales.sort_values(by="Total_Sales", ascending=False)

daily_sales
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = sales_pivot_sum.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales.sort_values(by="Total_Sales", ascending=False)

weekly_sales
```
